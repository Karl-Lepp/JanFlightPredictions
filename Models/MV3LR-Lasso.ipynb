```python
import numpy as np
import pandas as pd
```

```python
v3 = pd.read_csv("dfversion3.csv")
```

```python
v3
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import ElasticNetCV
```

```python
X = v3.drop(columns=['avg_arr_delay'])
y = v3['avg_arr_delay']
```

```python
print(type(X))
print(type(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
elastic = ElasticNetCV(l1_ratio=[.1, .2, .37, .4, .5, .7, .9, .95, .99], max_iter=10000)
```

```python
elastic.fit(X_train, y_train)
```

```python
y_pred = elastic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Linear Regression Score: ",elastic.score(X_test, y_test))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.feature_selection import SelectFromModel
v3.columns
```

```python
importance = np.abs(elastic.coef_)
feature_names = np.array(v3.drop(['avg_arr_delay'], axis=1).columns)
```

```python
threshold = np.sort(importance)[-51]
```

```python
sfm = SelectFromModel(elastic, threshold=threshold).fit(X_train, y_train)
```

```python
print(f"Features selected by SelectFromModel: {feature_names[sfm.get_support()]}")
```

```python
ag = ag.rename(columns = {0:'aircraft_group_0',
                     1:'aircraft_group_1',
                     2:'aircraft_group_2',
                     3:'aircraft_group_3',
                     4:'aircraft_group_4',
                     5:'aircraft_group_5',
                     6:'aircraft_group_6',
                     7:'aircraft_group_7',
                     8:'aircraft_group_8'})
```

```python
at = pd.get_dummies(pas["aircraft_type"], )
```

```python
cols = list(at.columns)
at.rename(columns={cols[i]:'aircraft_type_'+str(cols[i]) for i in range(len(cols)) } , inplace=True)
```

```python
at
```

```python
uc = pd.get_dummies(pas["unique_carrier"])
```

```python
cols = list(uc.columns)
uc.rename(columns={cols[i]:'unique_carrier_'+str(cols[i]) for i in range(len(cols)) } , inplace=True)
```

```python
re = pd.get_dummies(pas["region"])
```

```python
cols = list(re.columns)
re.rename(columns={cols[i]:'region_'+str(cols[i]) for i in range(len(cols)) } , inplace=True)
```
